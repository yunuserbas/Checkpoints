# API (45 min environ)

## 3.1 API adresses
Le jeu de données suivant recense une sélection des meilleurs restaurants de Paris, à prix très abordable (moins de 15 euros le menu en moyenne).  
En tant que Data Analyst en herbe, tu vas utiliser l'API [suivante](https://adresse.data.gouv.fr/api-doc/adresse), pour récupérer les coordonnées de ces restaurants.

```
food_paris = pd.read_csv("https://raw.githubusercontent.com/Gierko/food_paris_under_15_euros/main/food.csv")
```

```
food_paris
```

```
food_paris['whole_address'] = food_paris['adresse'] + ", " + food_paris['code postal']
food_paris['whole_address']
```

```
def API_adresse(adresse_postale):
  link_main = 'https://api-adresse.data.gouv.fr/search/?q='
  link = link_main + adresse_postale.replace(' ','+')
  return (link)
```

```
food_paris['API'] = food_paris['whole_address'].apply(API_adresse)
food_paris
```

```
import requests

def text(link_api):
  m = requests.get(link_api).json()
  point = m['features'][0]['geometry']['coordinates'][::-1]
  return (point)
```

```
food_paris['text'] = food_paris['API'].apply(text)
food_paris
```


## 3.2  Folium
Tu vas maintenant projeter ces restaurants sur une carte, en utilisant la librairie `Folium`.


```
import folium

coord = food_paris['text'][0]
m = folium.Map(location=coord, zoom_start = 13)

def geo(food_paris):
  for i in range(len(food_paris['text'])):    
    point=food_paris['text'][i]    
    folium.Marker(location=point,popup=food_paris['nom'][i]).add_to(m)
    

geo(food_paris)
m 
```
